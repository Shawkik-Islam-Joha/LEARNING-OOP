Inheritance in C++ is a fundamental object-oriented programming (OOP) concept that allows a class (called a derived or child class) to inherit properties and behaviors (data members and member functions) from another class (called a base or parent class). It promotes code reusability, establishes a relationship between classes, and enables polymorphism. Let me break it down step-by-step:

---

### Key Concepts of Inheritance
1. **Base Class (Parent Class)**: The class whose properties and methods are inherited by another class.
2. **Derived Class (Child Class)**: The class that inherits from the base class and can add its own properties or override inherited ones.
3. **Syntax**:
```cpp
class Base {
public:
    // Base class members
};

class Derived : access_specifier Base {
public:
    // Derived class members
};
```
- `access_specifier` can be `public`, `private`, or `protected`. It determines how the inherited members are accessible in the derived class.

---

### Types of Inheritance
1. **Single Inheritance**: A derived class inherits from one base class.
```cpp
class Animal {
public:
    void eat() { cout << "Eating...\n"; }
};

class Dog : public Animal {
public:
    void bark() { cout << "Barking...\n"; }
};

int main() {
    Dog d;
    d.eat();  // Inherited from Animal
    d.bark(); // Defined in Dog
    return 0;
}
```

2. **Multiple Inheritance**: A derived class inherits from more than one base class.
```cpp
class Father {
public:
    void height() { cout << "Tall\n"; }
};

class Mother {
public:
    void hair() { cout << "Curly\n"; }
};

class Child : public Father, public Mother {
};

int main() {
    Child c;
    c.height(); // From Father
    c.hair();   // From Mother
    return 0;
}
```

3. **Multilevel Inheritance**: A class is derived from a class that is itself derived from another class (a chain of inheritance).
```cpp
class Animal {
public:
    void eat() { cout << "Eating...\n"; }
};

class Mammal : public Animal {
public:
    void walk() { cout << "Walking...\n"; }
};

class Dog : public Mammal {
public:
    void bark() { cout << "Barking...\n"; }
};

int main() {
    Dog d;
    d.eat();  // From Animal
    d.walk(); // From Mammal
    d.bark(); // From Dog
    return 0;
}
```

4. **Hierarchical Inheritance**: Multiple derived classes inherit from a single base class.
```cpp
class Vehicle {
public:
    void start() { cout << "Starting...\n"; }
};

class Car : public Vehicle {
public:
    void drive() { cout << "Driving...\n"; }
};

class Bike : public Vehicle {
public:
    void ride() { cout << "Riding...\n"; }
};
```

5. **Hybrid Inheritance**: A combination of two or more types of inheritance (e.g., multiple and multilevel). This can lead to the "diamond problem" (explained later).

---

### Access Specifiers
- **`public`**: Inherited members retain their original access level (public remains public, protected remains protected).
- **`protected`**: Public and protected members of the base class become protected in the derived class.
- **`private`**: All members of the base class become private in the derived class (rarely used).

| Base Class Access | Public Inheritance | Protected Inheritance | Private Inheritance |
|-------------------|--------------------|-----------------------|---------------------|
| Public            | Public            | Protected             | Private             |
| Protected         | Protected         | Protected             | Private             |
| Private           | Not accessible    | Not accessible        | Not accessible      |

### The Diamond Problem
In multiple inheritance, if a class inherits from two classes that both inherit from a common base class, ambiguity arises. This is called the diamond problem.
```cpp
class A {
public:
    void show() { cout << "Class A\n"; }
};

class B : public A {};
class C : public A {};
class D : public B, public C {};

int main() {
    D d;
    d.show(); // Error: Ambiguous call to show()
    return 0;
}
```
**Solution**: Use the `virtual` keyword to ensure only one copy of the base class is inherited.
```cpp
class A {
public:
    void show() { cout << "Class A\n"; }
};

class B : virtual public A {};
class C : virtual public A {};
class D : public B, public C {};

int main() {
    D d;
    d.show(); // Works fine now
    return 0;
}
```

### Key Features
1. **Constructor and Destructor Order**:
   - Constructors are called from base to derived.
   - Destructors are called from derived to base.
```cpp
class Base {
public:
    Base() { cout << "Base Constructor\n"; }
    ~Base() { cout << "Base Destructor\n"; }
};

class Derived : public Base {
public:
    Derived() { cout << "Derived Constructor\n"; }
    ~Derived() { cout << "Derived Destructor\n"; }
};

int main() {
    Derived d;
    return 0;
}
// Output:
// Base Constructor
// Derived Constructor
// Derived Destructor
// Base Destructor
```

2. **Overriding**: A derived class can redefine a base class function.
```cpp
class Animal {
public:
    void sound() { cout << "Some sound\n"; }
};

class Dog : public Animal {
public:
    void sound() { cout << "Woof\n"; }
};
```

3. **Accessing Base Class Members**: Use the scope resolution operator (`::`) if needed.
```cpp
class Dog : public Animal {
public:
    void makeSound() {
        Animal::sound(); // Calls base class version
    }
};
```

---

### Benefits of Inheritance
- **Code Reusability**: Avoid rewriting common code.
- **Extensibility**: Easily extend functionality.
- **Polymorphism**: Use base class pointers/references to work with derived class objects.

### Limitations
- Tight coupling between classes.
- Complexity increases with multiple or hybrid inheritance.
